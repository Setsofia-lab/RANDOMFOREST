```python
import random
import pandas as pd
import numpy as np

df = pd.read_csv('/Users/samuelsetsofia/dev/projects/randomForest/train.csv')
df.head()
```

```python
df.dropna(inplace=True)
df
```

```python
features = ['Pclass','Sex','Age','SibSp','Parch', 'Fare', 'Embarked']
nb_train = int(np.floor(0.9 * len(df)))
df = df.sample(frac=1, random_state=217)
X_train = df[features][:nb_train]
y_train = df['Survived'][:nb_train].values
X_test = df[features][nb_train:]
y_test = df['Survived'][nb_train:].values
print(f'Traing data:{X_train.head()} , Testing data:{X_test.head()}, Traing labes: {y_train[5:]}, Testing labels : {y_test[5:]}')
```
